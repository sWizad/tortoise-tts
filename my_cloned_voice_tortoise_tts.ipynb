```python

import torch
import torchaudio
import torch.nn as nn
import torch.nn.functional as F

import IPython

from tortoise.api import TextToSpeech
from tortoise.utils.audio import load_audio, load_voice, load_voices

tts = TextToSpeech()

import os
from google.colab import files
```

# Run

```python
# upload at least 2 audio clips. They must be a WAV file, 6-10 seconds long.
CUSTOM_VOICE_NAME = "teng"

# Generate speech
voice_samples, conditioning_latents = load_voice(CUSTOM_VOICE_NAME)
```

## Generate Audio File (clone voice) 👌

```python
# Generate speech
kwargs = {'temperature': .8, 'length_penalty': 1.0, 'repetition_penalty': 2.0,
                'top_p': .8,
                'cond_free_k': 2.0, 'diffusion_temperature': 1.0,
                'conditioning_latents':conditioning_latents, 'voice_samples': voice_samples,
                'num_autoregressive_samples': 96, 'diffusion_iterations': 20,
                'method_name': 'phvb', 'method_order' : 2.5,
                }

text = """
In the wild world of mathematics, there's this concept called the Stability Region. It's like the safe zone for numerical method.
If the product between a step size delta and an eigenvalue lambda decides to step outside the Stability Region, well, the numerical solution are guarantee to diverge.
"""


gen = tts.tts(text, **kwargs)
out_path = f'generated-{CUSTOM_VOICE_NAME}.wav'
torchaudio.save(out_path, gen.squeeze(0).cpu(), 24000)
IPython.display.Audio(out_path)
```

# Upload new voice
